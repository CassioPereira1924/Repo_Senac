```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_dias = pd.DataFrame({'dias': ['Segunda', 'Tera', 'Quarta', 'Quinta', 'Sexta'],
           'Venda': [500, 200, 1000, 150, 1950]})

df_produto = pd.DataFrame({'Produto': ['PO', 'QUEIJO', 'PRESUNTO'],
                            'Vendas': [120, 90, 150]})

# Cria uma figura com 4 reas de grficos (2 linhas x 2 colunas)
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
# axs  uma matriz 2x2 de objetos Axes, onde cada posio representa um grfico especfico:
# axs[0, 0]  linha 0, coluna 0  canto superior esquerdo
# axs[0, 1]  linha 0, coluna 1  canto superior direito
# axs[1, 0]  linha 1, coluna 0  canto inferior esquerdo
# axs[1, 1]  linha 1, coluna 1  canto inferior direito

# Grfico 1: Linha - Vendas por dia da semana
axs[0, 0].plot(df_dias['dias'], df_dias['Venda'], marker='o', linestyle='-', color ='#FF8C00')
# axs[0, 0] acessa o primeiro grfico (linha 0, coluna 0)
# .plot() desenha uma linha conectando os valores de venda por dia
axs[0, 0].set_title('Vendas por Dia da Semana')   # Ttulo do grfico
axs[0, 0].set_xlabel('Dia da Semana')    # Rtulo do eixo X
axs[0, 0].set_ylabel('Valor Vendido')    # Rtulo do eixo Y

# Grfico 2: Barras verticais - Vendas por produto
axs[0, 1].bar(df_produto['Produto'], df_produto['Vendas'], color='green')
# axs[0, 1] acessa o segundo grfico (linha 0, coluna 1)
# .bar() desenha barras verticais para cada produto
axs[0, 1].set_title('Vendas por Produto')       # Ttulo do grfico
axs[0, 1].set_xlabel('Produto')                 # Rtulo do eixo X
axs[0, 1].set_ylabel('Vendas')                  # Rtulo do eixo Y

# Grfico 3: Barras horizontais - Vendas por produto
axs[1, 0].barh(df_produto['Produto'], df_produto['Vendas'], color='orange')
# axs[1, 0] acessa o terceiro grfico (linha 1, coluna 0)
# .barh() desenha barras horizontais para os produtos
axs[1, 0].set_title('Vendas por Produto (Horizontal)')  # Ttulo do grfico
axs[1, 0].set_xlabel('Vendas')                          # Rtulo do eixo X
axs[1, 0].set_ylabel('Produto')                         # Rtulo do eixo Y


# Grfico 4: Quadro de texto explicativo
axs[1, 1].axis('off')  # axs[1, 1] acessa o quarto grfico (linha 1, coluna 1)
# .axis('off') oculta os eixos para transformar o espao em um quadro de texto

# Insero de texto explicativo no grfico 4
axs[1, 1].text(0.1, 1, 'Resumo da Aula:', fontsize=14, color='black')
# .text(x, y, texto) insere texto na posio (10% largura, 100% altura)
# fontsize define o tamanho da fonte, color define a cor

axs[1, 1].text(0.1, 0.6, '- subplots usa matriz de eixos', fontsize=12)
# Explica que plt.subplots() retorna uma matriz de objetos Axes
# Cada grfico  acessado por sua posio: axs[linha, coluna]
axs[1, 1].text(0.1, 0.7, 'Senac Rio de Janeiro', fontsize=12)

axs[1, 1].text(0.1, 0.5, '- mais controle sobre cada grfico', fontsize=12)
# Destaca que podemos configurar cada grfico individualmente

axs[1, 1].text(0.1, 0.4, '- ideal para projetos mais complexos', fontsize=12)
# Mostra que subplots so teis para dashboards e relatrios com mltiplas visualizaes

# Ajusta os espaos entre os grficos para evitar sobreposio
plt.tight_layout()

# Exibe todos os grficos juntos
plt.show()
```
