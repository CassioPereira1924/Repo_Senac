```python
df = pd.read_csv('https://www.ispdados.rj.gov.br/Arquivos/BaseDPEvolucaoMensalCisp.csv', sep= ';', encoding= 'latin1')
df.head()
```

```python
furto_celular = df.groupby('cisp')['furto_celular'].sum().reset_index()
furto_celular
```

```python
array_furtos = np.array(furto_celular['furto_celular'])
mediana = np.median(array_furtos)
media = np.mean(array_furtos)
q1 = np.percentile(array_furtos, 25)
q2 = np.percentile(array_furtos, 50)
q3 = np.percentile(array_furtos, 75)
distancia = (media - mediana) / mediana
iqr = q3 - q1
limite_superior = q3 + (1.5 * iqr)
limite_inferior = q1 - (1.5 * iqr)
print(f'media {media}')
print(f'mediana {mediana}')
print(f'distância {distancia*100}%' )
```

```python
furtos_outliers = furto_celular.loc[furto_celular['furto_celular'] >= limite_superior]
furtos_outliers.sort_values(by="furto_celular", ascending= False)
```
